```python
import numpy as ns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read file from from path location
df = pd.read_csv('football_mini.csv')
```

```python
#Have a view of dataset few lines
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
#Plot of a histogram with players age
plt.figure(figsize=(7, 5))
plt.hist(df['Age'], bins=10, color='green', edgecolor='black', alpha=0.7)
#Axes labels and title
plt.title('Histogram of Players Ages', color='blue', fontsize=12, fontweight='bold')
plt.xlabel('Age Value')
plt.ylabel('Freqeuncy')
#saving the histogram image
plt.savefig('histogram of Players Age', dpi=96)
plt.show()
```

```python
#scatter plot of players Age and the thier finishig 
plt.figure(figsize=(6, 4))
plt.scatter(df['Age'], df['Finishing'], color='green', 
            edgecolor='green', alpha=0.7)
#Title and axes labeling with customizing
plt.xlabel('Age', color='black', fontweight='bold')
plt.title('Age vs Finishing', color='blue', fontsize=12, fontweight='bold' )
plt.ylabel('Finishing in %', color='black', fontsize=12, fontweight='bold')
#saving my image
plt.savefig('Age vs Finishing', dpi=96)
plt.show()
```
