```python
import numpy as ns
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Read file from from path location
df = pd.read_csv('football_mini.csv')
```

```python
#Have a view of dataset few lines
print(df.head())
```

```python
#Plot a simple histogram 
plt.hist(df['Age'], bins=12, color='green', edgecolor='black', alpha=0.7)
#Axes labels and title
plt.title('Histogram of Players Ages', color='blue', fontsize=12, fontweight='bold')
plt.xlabel('Age Value')
plt.ylabel('Freqeuncy')
plt.show()
```

```python
#scatter plot of players Age and the thier finishig 
plt.figure(figsize=(12,8))
plt.scatter(df['Age'], df['Finishing'], color='blue', 
            edgecolor='blue', alpha=0.8)
#Title and axes labeling with customizing
plt.xlabel('Age', color='black', fontweight='bold')
plt.title('Age vs Finishing', color='red', fontsize=12, fontweight='bold' )
plt.ylabel('Finishing', color='black', fontsize=12, fontweight='bold')
plt.show()
```
